**DATA UNNESTING PROCESS:**

To enhance the granularity of our analysis, we perform unnesting on key columns containing multiple values.

**Step 1 : Cast Column -** The cast column is split and unnested into individual entries using **str.split() and stack()**. This allows us to analyze
each cast member separately.

**Step 2 : Director Column -** Similarly, the director column is unnested to handle multiple directors associated with different titles.


**Step 3 : Genre Column -** The listed-in column is split and unnested to capture each genre individually for more detailed genre-based insights.

**Step 4 : Final DataFrame -** The original columns are dropped, and unnested data is combined into a refined dataFrame df_final.

This unnesting process enables precise analysis by treating individual actors, directors and genres as distinct entities.

```python
df_cast_split = netflix_data['cast'].str.split(',', expand=True)
df_cast_unnest = df_cast_split.stack().reset_index(level=1, drop=True).to_frame('cast')
df_director_split = netflix_data['director'].str.split(',', expand=True)
df_director_unnest = df_director_split.stack().reset_index(level=1, drop=True).to_frame('director')
df_listed_split = netflix_data['listed_in'].str.split(',', expand=True)
df_listed_unnest = df_listed_split.stack().reset_index(level=1, drop=True).to_frame('listed_in')
df_final = netflix_data.drop(['cast', 'director', 'listed_in'], axis=1)\
.join([df_cast_unnest, df_director_unnest, df_listed_unnest])

df_final
```

**Best Time to Watch a TV Show :**

To determine the best time to launch a TV show, we analyze historical data of TV shows added to the platform and then begin by ltering the
dataset to focus solely on TV shows and converting the 'date_added' column into datetime format. Next, we extracte the month and year,
ensuring the months are arranged chronologically for accurate plotting and then group the data by title to count unique shows added each
month and visualize this information with a bar graph. Additionally, we calculate yearly trends by counting unique titles added per year. A heatmap
is generated to showcase the distribution of TV shows across various years and months. By analyzing these visualizations, we can pinpoint
optimal release periods for new shows, facilitating strategic planning. This data-driven approach enhances our ability to launch TV shows at
times that maximize audience engagement.


**Insights :** The data spans from 2012 to 2021, with each cell representing a specic month and year. The color scale ranges from light yellow
(fewer shows) to dark blue (more shows).


**Business Insights :** The most notable peak is in July 2021, where 88 unique TV shows were added, the highest in the dataset.

**Recommendations :** :This suggests a significant increase in content addition during that period. Overall, the heatmap reveals seasonal trends,
with certain months consistently showing higher additions of TV shows over the years.


```python
tv_shows = df_final[df_final['type'] == 'TV Show']
tv_shows['date_added'] = pd.to_datetime(tv_shows['date_added'].str.strip(), errors='coerce')
tv_shows['month'] = tv_shows['date_added'].dt.month_name()
tv_shows['year'] = tv_shows['date_added'].dt.year
months_order = ['January', 'February', 'March', 'April', 'May', 'June',
'July', 'August', 'September', 'October', 'November', 'December']
title_counts = tv_shows.groupby('title').size()
monthly_counts = tv_shows.groupby('month')['title'].nunique().reindex(months_order, fill_value=0)
yearly_counts = tv_shows.groupby('year')['title'].nunique()

# Bar chart
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Unique TV Shows Added by Month')
plt.xlabel('Month')
plt.ylabel('Number of Unique TV Shows')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Heatmap
heatmap_data = tv_shows.pivot_table(index='year', columns='month', values='title', aggfunc='nunique')
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='YlGnBu', cbar_kws={'label': 'Number of Unique TV Shows'})
plt.title('Heatmap of Unique TV Shows Added Over Years by Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.tight_layout()
plt.show()
```

Stand-Up Comedy & Talk Shows. **Action & Adventure :** Action & Adventure, TV Action & Adventure, Thrillers. **Documentary :** Documentaries,
Docuseries. **Horror & Thriller :** Horror Movies, TV Horror, TV Mysteries, TV Thrillers. **Kids & Family :** Children & Family Movies, Kids' TV. **Sci-Fi &
Fantasy :** Sci-Fi & Fantasy, TV Sci-Fi & Fantasy. **International :** International Movies, International TV.   **Shows Others :** All remaining genres

**Business Insights :** Other countries should expand their genre range, as this would increase the volume of available content across countries and cater to a wider range of audience preference.

**Recommendations :** Netflix should diversify its content by adding various genres, particularly in the countries with limited range, in order to enhance the viewers engagement and explore different audience preference.

```python
# Define genre grouping

genre_mapping = {
'Dramas': 'Drama',
'Romantic TV Shows': 'Drama',
'Romantic Movies': 'Drama',
'TV Dramas': 'Drama',
'Comedies': 'Comedy',
'Stand-Up Comedy': 'Comedy',
'TV Comedies': 'Comedy',
'Stand-Up Comedy & Talk Shows': 'Comedy',
'Action & Adventure': 'Action & Adventure',
'TV Action & Adventure': 'Action & Adventure',
'Thrillers': 'Action & Adventure',
'Documentaries': 'Documentary',
'Docuseries': 'Documentary',
'Horror Movies': 'Horror & Thriller',
'TV Horror': 'Horror & Thriller',
'TV Mysteries': 'Horror & Thriller',
'TV Thrillers': 'Horror & Thriller',
'Children & Family Movies': 'Kids & Family',
'Kids\' TV': 'Kids & Family',
'Sci-Fi & Fantasy': 'Sci-Fi & Fantasy',
'TV Sci-Fi & Fantasy': 'Sci-Fi & Fantasy',
'International Movies': 'International',
'International TV Shows': 'International',
}


top_countries = df_final['country'].value_counts().head(10).index
df_top_countries = df_final[df_final['country'].isin(top_countries)]
df_top_countries['listed_in'] = df_top_countries['listed_in'].map(genre_mapping)
country_genre_counts = df_top_countries.groupby(['country', 'listed_in'])['title'].nunique().unstack(fill_value=0)

# Plotting a stacked bar chart

country_genre_counts.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')
plt.title('Content Distribution by Genre in Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Unique Titles')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
